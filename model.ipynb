# Imports 


```python
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
import numpy as np
```

# Load Data 

```python
df = pd.read_csv("weather_last_30_days.csv")
df ["date"]= pd.to_datetime(df["date"])

print(df.head())
```

```python
missing=df.isnull().mean()*100
print (missing)
```

# create Features 
use the previous days temparature to predict the next day temparature.

```python
df ["temp_next_day"]=df["temp_max"].shift(-1)

df=df.dropna()
```

```python
print (df)
```

# Train Model 

```python
x= df [["temp_max", "temp_min", "rain_mm"]]
y= df ["temp_next_day"]

x_train , x_test , y_train , y_test = train_test_split(x , y , test_size=0.2,random_state=42)
model = RandomForestRegressor(
    n_estimators=100,
    random_state=42
)
model.fit(x_train,y_train)
```

# Evaluate Model


```python
pred = model.predict(x_test)

mae = mean_absolute_error(y_test,pred)

print ("MAE:",mae)
```

# Predict Next 7 Days 

```python
last_row = x.iloc[-1].copy()

future = []

for day in range(7):
    # prepare one-row DataFrame for prediction
    input_df = pd.DataFrame([
        [last_row['temp_max'], last_row['temp_min'], last_row['rain_mm']]
    ], columns=['temp_max', 'temp_min', 'rain_mm'])

    # predict next day's max temperature
    prediction = model.predict(input_df)[0]

    future.append(prediction)

    # update last_row for next iteration (keep other features unchanged)
    last_row['temp_max'] = prediction

print(future)
for day, temp in enumerate(future, start=1):
    print(f"Day {day}: {temp:.2f}°C")
```
